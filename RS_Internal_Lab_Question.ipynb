### Q1. Read the data (jokes.csv)

```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt 
```

```python
#Loading the dataset
data_jokes = pd.read_csv("jokes.csv")
```

### Q2. Create new Dataframe `ratings` starting from column `joke1` and consider first 200 rows only



```python
#Creating a new dataframe starting from column joke 1 
ratings = pd.DataFrame(data_jokes.iloc[:200,1:])
ratings
```

### Q3. Preprocess `ratings`


*   Verify presence of null values
*   Replace null with 0



```python
#As mentioned above 99 means null value 
#Finding 99 and replacing it with 0 in ratings data

ratings.isnull().any().sum()
#No null values 
```

```python
#Finding 99 and replacing it with 0 in ratings data
ratings.replace(99,0,inplace=True)
#eyeballing the dataset , to check for 99
ratings
#all 99 is replaced with 0.00
```

### Q4. Find mean for each joke


*   Find mean of 100 jokes



```python
col_name='average_rating'
mean_rating = pd.DataFrame(ratings.mean())
mean_rating
```

### Q5. List most popular 10 jokes

```python
#Popularity based model 
mean_rating.sort_values(0,ascending=False).head(10)
#Joke 50 got mean highest rating of 4.05
#Hence can be considered as popular jokes
```
